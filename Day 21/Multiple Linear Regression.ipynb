```python
x = data.iloc[:, :1]
y = data.iloc[:, 1:]
```

```python
from sklearn.model_selection import train_test_split
y_train, y_test, x_train, x_test = train_test_split(y ,x, test_size = 0.2, random_state = 2)
```

```python
from sklearn.linear_model import LinearRegression
multiple_regression = LinearRegression()
```

```python
multiple_regression .fit(y_train, x_train)
```

```python
multiple_regression.coef_
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
l = multiple_regression.predict(y_test)
```

```python
#Root mean square error (RMSE) is a method of measuring the difference between values predicted by a model and their actual values.
RMSE = np.sqrt(mean_squared_error(x_test, l))
r_square = r2_score(x_test, l)

print('The R^2 value is:', r_square)
print('The RMSE value is:', RMSE)
```
